```python
df['Calories (% Daily Value)'] = 'default'
for x in df.index:
    perDay = healthyAverageCalories
    calories = float(df.loc[x,'Calories'])
    if(calories > 0):
        percentage = round((calories * 100) / perDay,1)
        df.loc[x,'Calories (% Daily Value)'] = percentage
    elif(calories == 0):
        df.loc[x,'Calories (% Daily Value)'] = 0


df['Trans Fat (% Daily Value)'] = 'default'
for x in df.index:
    perDay = healthyAverageTransFat
    transFat = float(df.loc[x,'Trans Fat'])
    if(transFat > 0):
        percentage = round((transFat * 100) / perDay,1)
        df.loc[x,'Trans Fat (% Daily Value)'] = percentage
    elif(transFat == 0):
        df.loc[x,'Trans Fat (% Daily Value)'] = 0


df['Protein (% Daily Value)'] = 'default'
for x in df.index:
    perDay = healthyAverageProtein
    protein = float(df.loc[x,'Protein'])
    if(protein > 0):
        percentage = round((protein * 100) / perDay,1)
        df.loc[x,'Protein (% Daily Value)'] = percentage
    elif(protein == 0):
        df.loc[x,'Protein (% Daily Value)'] = 0


df['Sugars (% Daily Value)'] = 'default'
for x in df.index:
    perDay = healthyAverageSugars
    sugars = float(df.loc[x,'Sugars'])
    if(sugars > 0):
        percentage = round((sugars * 100) / perDay,1)
        df.loc[x,'Sugars (% Daily Value)'] = percentage
    elif(sugars == 0):
        df.loc[x,'Sugars (% Daily Value)'] = 0
```

```python
df['Proportion of Carbohydrates (%)'] = 'default'
for x in df.index:
    total = float(df.loc[x,'Total Fat']) + float(df.loc[x,'Carbohydrates']) + float(df.loc[x,'Protein'])
    carbohydrates = float(df.loc[x,'Carbohydrates'])
    if(carbohydrates > 0):
        proportion = round((carbohydrates * 100) / total,1)
        df.loc[x,'Proportion of Carbohydrates (%)'] = proportion
    elif(carbohydrates == 0):
        df.loc[x,'Proportion of Carbohydrates (%)'] = 0
        

df['Proportion of Protein (%)'] = 'default'
for x in df.index:
    total = float(df.loc[x,'Total Fat']) + float(df.loc[x,'Carbohydrates']) + float(df.loc[x,'Protein'])
    protein = float(df.loc[x,'Protein'])
    if(protein > 0):
        proportion = round((protein * 100) / total,1)
        df.loc[x,'Proportion of Protein (%)'] = proportion
    elif(protein == 0):
        df.loc[x,'Proportion of Protein (%)'] = 0


df['Proportion of Fat (%)'] = 'default'
for x in df.index:
    total = float(df.loc[x,'Total Fat']) + float(df.loc[x,'Carbohydrates']) + float(df.loc[x,'Protein'])
    fat = float(df.loc[x,'Total Fat'])
    if(fat > 0):
        proportion = round((fat * 100) / total,1)
        df.loc[x,'Proportion of Fat (%)'] = proportion
    elif(fat == 0):
        df.loc[x,'Proportion of Fat (%)'] = 0
       

df['Saturated Fat Percentage in Fat'] = 'default'
for x in df.index:
    fat = float(df.loc[x,'Total Fat'])
    saturated = float(df.loc[x,'Saturated Fat'])
    if(fat > 0):
        proportion = round((saturated * 100) / fat,1)
        df.loc[x,'Saturated Fat Percentage in Fat'] = proportion
    elif(fat == 0):
        df.loc[x,'Saturated Fat Percentage in Fat'] = 0

count = 1
for x in df.index:
    carbohydrates = df.loc[x,'Proportion of Carbohydrates (%)']
    protein = df.loc[x,'Proportion of Protein (%)']
    fat = df.loc[x,'Proportion of Fat (%)']
    saturated = df.loc[x,'Saturated Fat Percentage in Fat']

    #when adding the condition that saturated <= 11
    
    
    saturated = df.loc[x,'Saturated Fat Percentage in Fat']
    
    #Assuming that an average is three meals a day,
    #we look that the nutrient % Daily Value is not more than (100/3)
    if(45 <= carbohydrates <= 65 and 10 <= protein <= 35 and  
       20 <= fat <= 35 and float(df.loc[x,'Carbohydrates (% Daily Value)']) <= (100/3)
       and float(df.loc[x,'Total Fat (% Daily Value)']) <= (100/3)
       and float(df.loc[x,'Saturated Fat (% Daily Value)']) <= (100/3)
       and float(df.loc[x,'Protein (% Daily Value)']) <= (100/3)):
        
        print(str(count) + ". " + df.loc[x,'Category'] + ": " + str(df.loc[x,'Item']))
        count = count + 1
```
